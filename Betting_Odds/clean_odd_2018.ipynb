```python
from path import Path
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
```

```python
csv_path = Path('mlb_odds_2018.csv')
df= pd.read_csv(csv_path)
df.head(10)
```

```python
df= df[["Date", "VH", "Team", "Pitcher","Open", "Close"]]
```

```python
df["Date"]= df["Date"].astype(str).apply(lambda x: x[:-2] + "-" + x[-2:] + "-2019")
```

```python
df["Date"]= pd.to_datetime(df["Date"])
```

```python
df.head()
```

```python
df["Pitcher"]= df["Pitcher"].astype(str).apply(lambda x: x[:-2])
```

```python
df["Close"]= df["Close"].astype(int)
```

```python
df.to_csv("mlb_odds_2018.csv")
```
