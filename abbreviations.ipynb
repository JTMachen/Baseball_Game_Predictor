# Abbreviations

R: Runs.

RA: Run Average.

Opp: Opponent.

Inn: Innings.

W/L: Win/Lose.

W-L: Win - Lose.

GB: Games back. How far the team is from the leader of the division. The formula is ((Division Leader's Wins - This team's Wins) + (This team's Losses - Division Leader's Losses)) / 2.

G: Games played by the player.

AB: At Bat. Different than plate appearance. When a batter reaches first or other bases, or when he scores a sacrifice fly.

PA: Plate Appearance. When a player completes a turn batting.

H: when a player reaches a base due strictly to the streak of the ball with the bat, and not due to error of the pitcher.

1B, 2B, 3B, HR: when the player reaches 1, 2, 3, or Home Run due strictly to a hit.

RBI: number of runners who score due to a batters' action, except when batter grounded into double play or reached on an error.

BB: Base on Balls. hitter not swinging at four pitches called out of the strike zone and awarded first base.

SF: Sacrifice Fly. When a batter and a teammate score a run even though the ball was caught in the fly for an out for the batter.

SH: number of sacrifice bunts which allow runners to advance on the basepaths.

GDP: number of ground balls hit that became double plays.

SB: Stolen Base.

CA:  Caught stealing: times tagged out while attempting to steal a base.

BA:  Batting Avarage. Also AVG, hits divided by at bats (H/AB).

HBP: Hit by pitch. times touched by a pitch and awarded first base as a result.

OBP:  On-base percentage: times reached base (H + BB + HBP) divided by at bats plus walks plus hit by pitch plus sacrifice flies (AB + BB + HBP + SF).

SLG: slugging percentage is a measure of the batting productivity of a hitter. It is calculated as total bases divided by at bats. Unlike batting average, slugging percentage gives more weight to extra-base hits with doubles, triples, and home runs, relative to singles. Walks are specifically excluded from this calculation, as a plate appearance that ends in a walk is not counted as an at bat.

OPS: On-base plus slugging. formula: OBP + SLG.

AVG: same as BA.

FB: Fly Balls. A type of batted ball where the ball is hit high into the air. Players in the field attempt to catch the ball before it hits the ground to get the batter out automatically. Also referred to as a "fly".

LD: Line Drive. Also known as a liner, a line drive is a batted ball that is hit hard in the air and has a low arc.

IFFB: infield fly ball.

IFH: infield hit, which is the percentage of ground balls a player hits that end up being infield hits.

BU: number of bunts.

BUH: Bunt Hits. The number of hits a batter achieved by bunting the ball.

K: Strike out (also abbreviated SO). number of times that a third strike is taken or swung at and missed, or bunted foul. Catcher must catch the third strike or batter may attempt to run to first base.

ISO: Isolated Power. a hitter's ability to hit for extra bases, calculated by subtracting batting average from slugging percentage.
